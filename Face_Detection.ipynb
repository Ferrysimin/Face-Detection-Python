```python
#menampilkan vidio
import cv2
import numpy as np

#load Video
cap = cv2.VideoCapture('video.mp4')

#interate for each frame in vidio
while(cap.isOpened()):
    #get image on each frame 
    ret, frame = cap.read()
    if ret == True:
        #show image
        cv2.imshow('Frame', frame)
        
        #closing using 'q'
        if cv2.waitKey(25) & 0xFF == ord('q') :
            break
            
    else:
        break
            
#close vidio
cap.release()
#close window
cv2.destroyAllWindows()
```

```python
#vidi dan disimpan
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 5, (640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        out.write(frame)

        cv2.imshow('frame',frame)
        if cv2.waitKey(5) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()

out.release()

cv2.destroyAllWindows()
```

```python
#Deteksi mata & mulut
import numpy as np
import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
 
img = cv2.imread("alam.jpg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x, y, w, h) in faces:
    img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    rol_gray = gray[y:y+h, x:x+w]
    rol_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(rol_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(rol_color, (ex, ey), (ex+ew, ey+eh), (255, 0, 0), 2)
        
cv2.imshow('img', img)
        
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Face detection Video
import cv2


face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

 
cap = cv2.VideoCapture(0)

#cap = cv2.VideoCapture('filename.mp4')

while True:
    _, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    cv2.imshow('img', img)
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
